```python
import numpy as np
import pandas as pd
```

```python
btc = pd.read_csv('BTC_day.csv')
btc.drop('Unnamed: 0', axis=1, inplace=True)
btc
```

```python
from datetime import datetime as dt

btc['year'] = btc['OpenTime'].apply(lambda x: dt.fromtimestamp(x / 1000).year)
btc['month'] = btc['OpenTime'].apply(lambda x: dt.fromtimestamp(x / 1000).month)
btc['day'] = btc['OpenTime'].apply(lambda x: dt.fromtimestamp(x / 1000).day)
btc['weekday'] = btc['OpenTime'].apply(lambda x: dt.fromtimestamp(x / 1000).weekday())
```

```python
btc.drop(columns=['bav', 'OpenTime', 'CloseTime', 'tbv', 'tbBav', 'ignore'], inplace=True)
btc
```

```python
import plotly.graph_objects as go

def c_chart(data,label):
    candlestick = go.Figure(data = [go.Candlestick(x=data.index,
                                                   open = data['open'], 
                                                   high = data['high'], 
                                                   low = data['low'], 
                                                   close = data['close'])])
    candlestick.update_xaxes(title_text = 'Time',
                             rangeslider_visible = True)

    candlestick.update_layout(
    title = {
            'text': '{:} Candelstick Chart'.format(label),
            "y":0.8,
            "x":0.5,
            'xanchor': 'center',
            'yanchor': 'top'})

    candlestick.update_yaxes(title_text = 'Price in USD', ticksuffix = '$')
    return candlestick
```

```python
%matplotlib inline
btc_candle=c_chart(btc, label="BTC Price")
btc_candle.show()
```

```python
import ta
from ta.momentum import roc, rsi, stoch
from ta.trend import macd, adx, sma_indicator, PSARIndicator
from ta.volume import on_balance_volume, volume_price_trend, money_flow_index

def indicators(df):
    # Calculate moving averages
    df['ma_20'] = ta.trend.SMAIndicator(df['close'], window=5).sma_indicator()
    df['ma_50'] = ta.trend.SMAIndicator(df['close'], window=10).sma_indicator()

    # Calculate relative strength index (RSI)
    df['rsi'] = ta.momentum.RSIIndicator(df['close'], window=14).rsi()

    # Calculate moving average convergence divergence (MACD)
    macd = ta.trend.MACD(df['close'], window_slow=26, window_fast=12, window_sign=9)
    df['macd'] = macd.macd()
    df['macd_signal'] = macd.macd_signal()

    # Calculate Bollinger Bands
    bb = ta.volatility.BollingerBands(df['close'], window=20, window_dev=2)
    df['bb_high'] = bb.bollinger_hband()
    df['bb_low'] = bb.bollinger_lband()

    # Calculate Rate of Change (ROC)
    df['roc'] = ta.momentum.ROCIndicator(df['close'], window=12).roc()
    
    return df
```

```python
btc = indicators(btc)
btc.head(50)
```
